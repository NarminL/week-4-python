```python
# -----------------------------------------------------------------
# Decision Tree Classifier
# Predict the income of an adult based on the census data
# -----------------------------------------------------------------

# Import libraries
import pandas as pd 
import numpy as np
```

```python
# Read dataset
dec=pd.read_csv(r'C:\Users\Sevinc\Downloads\decisiontreeAdultIncome.csv')
decPrep= dec.copy()
decPrep
```

```python
# Check for Null values
dec.isnull()
```

```python
# Create Dummy variables
decPrep.dtypes
decPrep = pd.get_dummies(decPrep, drop_first=True)
decPrep
```

```python
# Create X and Y Variables
Y = decPrep[['race_ Black']]
X = decPrep.drop(['race_ Black'], axis=1)
print(X)
print(Y)
```

```python
# Split the X and Y dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = \
train_test_split(X, Y, test_size = 0.3, random_state = 1234, stratify=Y)
```

```python
# Import and train classifier
from sklearn.tree import DecisionTreeClassifier
decision_clf = DecisionTreeClassifier()
decision_clf = decision_clf.fit(X_train,Y_train)
```

```python
# Test the model
from sklearn.metrics import accuracy_score
clf = DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X_train,Y_train)

y_pred = clf.predict(X_test)
print("Accuracy:",accuracy_score(Y_test, y_pred))
```

```python
# Evaluate the model
from sklearn.metrics import classification_report

print(classification_report(Y_test,y_pred))
```
